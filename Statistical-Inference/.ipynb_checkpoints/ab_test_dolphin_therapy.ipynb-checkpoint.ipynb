```python
from datascience import *
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plots
plots.style.use('fivethirtyeight')
```

```python
therapy = np.repeat(['Dolphin', 'No Dolphin'], [13, 12]) #make a list of 13 people swimming with dolphin and 12 not
therapy
```

```python
improvement = np.repeat([1, 0, 1, 0], [10, 3, 3, 9]) #for dolphin group, 10 show improvement, 3 do not. For no dolohin group, 3 show improvement, 9 don't
improvement
```

```python
dolphin = Table().with_columns('Therapy', therapy, 'Improvement', improvement)
dolphin
```

```python
dolphin_summary = dolphin.group('Therapy', np.mean) #proportions of improvement
dolphin_summary
```

```python
dolphin_summary.column(1).item(0) - dolphin_summary.column(1).item(1) #difference of improvement proportions
```

```python
def difference_of_props(table, outcome, group_label):
    """Takes: name of table, column label of binary categorical variable,
    column label of group-label variable
    Returns: Difference of proportions of the two groups"""
    
    #table with the two relevant columns
    reduced = table.select(group_label, outcome)
    
    # table containing counts for each group
    summary = reduced.group(group_label, np.mean)

    return summary.column(1).item(0) - summary.column(1).item(1)
    
    
```

```python
difference_of_props(dolphin, 'Improvement', 'Therapy')
```

```python
def one_simulated_difference(table, label, group_label):
    """Takes: name of table, column label of binary cateogorcial variable,
    column label of group-label variable
    Returns: Difference of proportions of the two groups after shuffling labels"""
    
    # array of shuffled labels
    shuffled_array = table.sample(with_replacement = False).column(group_label)
    
    # table of outcome variable and shuffled labels
    shuffled_table = table.select(label).with_column('Shuffled Label', shuffled_array)

    return difference_of_props(shuffled_table, label, 'Shuffled Label')
```

```python
one_simulated_difference(dolphin, 'Improvement', 'Therapy')
```

```python
difference = make_array()

for i in np.arange(10000):
    difference = np.append(difference, one_simulated_difference(dolphin, 'Improvement', 'Therapy'))
```

```python
Table().with_column('Differences in Group Proportions', difference).hist()
plots.axvline(difference_of_props(dolphin, 'Improvement', 'Therapy'), color = 'red', lw = 2)
```

```python
np.mean(difference >= difference_of_props(dolphin, 'Improvement', 'Therapy'))
```
